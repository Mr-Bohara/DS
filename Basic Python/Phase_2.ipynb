```python
# chapter 5 - Loops

# For loop
for i in range(1, 3):
    print(i)
for i in range(4, 0, -1):
    print(i)

total = 0
for number in range(1, 101):
    total += number
print(f"sum of 1 to 100: {total}")

n = 7
print(f"nultiplication of {n}: ")
for i in range (1, 11):
    print(f"{n} * {n * i}")

print("even numbers: ")
for num in range(2, 21, 2):
    print(num, end=" ")



#while loop
print("While Loop use")

count = 1
while count <=5:
    print(f"count is: {count}")
    count += 1
print("loop finished.")

password = ""

while password != "admin":
    password = input("enter password: ")
    if password != "admin":
        print("error")
print("login successfull")

secret = 42
guess = 0
while guess != secret:
    guess = int(input("Guess the munber: "))
    if guess < secret:
        print("Too low")
    elif guess > secret:
        print("Too hugh!")
print("Correct! You yot it.")


#break and continue

for i in range(1, 6):
    if i == 4:
        break
    print(i)

for i in range(1, 6):
    if i % 2 == 0:
        continue
    print(i)
```
